```python
karte = gpd.read_file('/Users/susannekorbel/Desktop/BEZIRKSGRENZEOGD.json')
bezirke_daten = pd.read_excel(r'/Users/susannekorbel/Desktop/Bevoelkerungsstand1870.xlsx', sheet_name='bettgeher')
```

```python
karte
```

```python
m = folium.Map(location=[48.2, 16.3], zoom_start=12)
```

```python
m
```

```python
folium.Choropleth(
    geo_data=karte,
    name="choropleth",
    data=bezirke_daten,
    columns=["district", "percent"],
    key_on="feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Bedlodgers Rate (%)",
).add_to(m)

folium.LayerControl().add_to(m)
```
